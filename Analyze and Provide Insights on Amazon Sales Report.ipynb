```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from a local file
file_path = r"C:\Users\LENOVO\Desktop\tkproject.install\Amazon Sale Report.csv"
df = pd.read_csv(file_path)


# Display the first few rows of the dataset to inspect column names
print("Dataset Columns:", df.columns)
print(df.head())

# Filter out cancelled orders
filtered_data = df[df['Status'] != 'Cancelled']

# Group by 'Category' and sum the sales amount
category_sales = filtered_data.groupby('Category')['Amount'].sum().reset_index()

# Sort categories by total sales in descending order
popular_categories = category_sales.sort_values(by='Amount', ascending=False)

# Display popular categories
print("Most Popular Product Categories Based on Sales Volume:")
print(popular_categories)

# Bar Plot of Total Sales by Product Category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Amount', data=popular_categories)
plt.title('Total Sales by Product Category')
plt.xticks(rotation=45)  # Rotate x labels for readability
plt.ylabel('Total Sales Amount (INR)')
plt.xlabel('Product Category')
plt.tight_layout()
plt.show()
```
